```python
import matplotlib.pyplot as plt
import json
```

```python
data = None
with open('results400.json') as f:
    data = json.load(f)
for nn_key in data.keys():
    nn_data = data[nn_key]
    train_loss = nn_data["train_loss"]
    test_loss = nn_data["test_loss"]
    test_accuracy = nn_data["test_accuracy"]

    # print(f"NN: {nn_key}, max accuracy: {max(test_accuracy)}")
    # print(f"NN: {nn_key}, min test loss: {min(test_loss)}")
    plt.figure()
    epochs = len(train_loss)
    # plt.plot(range(1, epochs + 1), test_loss)
    plt.plot(range(1, epochs + 1), test_accuracy)
    # plt.legend(["Train loss", "Test loss"])
    plt.title(nn_key)
    plt.show()
```
